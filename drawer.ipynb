```python

def work(m=100, n=100, x_bounds=x_bounds, y_bounds=y_bounds, test=fil_test.values):
    epsila = 0.25
    x_grid = np.linspace(x_bounds[0] - epsila, x_bounds[1] + epsila, num=m + 1)
    y_grid = np.linspace(y_bounds[0] - epsila, y_bounds[1] + epsila, num=n + 1)

    x_shift, y_shift = x_grid[1] - x_grid[0], y_grid[1] - y_grid[0]
    print(x_shift, y_shift)
    dot_counter = np.zeros([m,n]) # granitsa
    colors = np.zeros([m,n])
    for tweet_index, tweet in enumerate(tqdm(test)):
        if(((x_bounds[0] <= tweet[10] <= x_bounds[1]) and (y_bounds[0] <= tweet[9] <= y_bounds[1])) == False):
            continue
        x_pos = (tweet[10] - x_bounds[0]) / x_shift
        y_pos = (tweet[9] - y_bounds[0]) / y_shift
        dot_counter[x_pos][y_pos] += 1.
        colors[x_pos][y_pos] += res[tweet_index][1]

    az = np.zeros([m,n])
    for i in range(m):
        for j in range(n):
            az[i][j] = colors[i][j] / dot_counter[i][j] if dot_counter[i][j] != 0 else 0.

    plt.figure(figsize=(20,20))
    plt.imshow(az)
    
    return az
    
```

```python
indices = random.sample(range(0, len(texts)), 100000)
tweets_dict_to_stemm = {}
for tweet in tqdm([texts[ind] for ind in indices]):
    for term in tweet.split():
        if term not in tweets_dict_to_stemm:
            tweets_dict_to_stemm[term] = stemmer(term)
```

```python
tweets_dict_to_stemm
```

```python
stemmed_docs = []
for doc in tqdm([texts[ind] for ind in indices]):
    to_push = []
    for term in doc:
        to_push.append(tweets_dict_to_stemm[term])
    stemmed_docs.append(to_push)
```
